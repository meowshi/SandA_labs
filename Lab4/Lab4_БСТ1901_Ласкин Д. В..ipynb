## Классы стека и дека

```python
# Класс стека
class Stack:
    def __init__(self, name = "NoName"):
        self.stack = []
        self.size = 0
        self.name = name
    def push(self, value):
        self.stack.append(value)
        self.size += 1
    def pop(self):
        if self.is_empty(): 
            return None
        self.size -= 1
        return self.stack.pop()
    def peek(self):
        return self.stack[self.size - 1]
    def is_empty(self):
        return self.size == 0
    def getName(self):
        return self.name
# Класс дека
class Deque:
    def __init__(self):
        self.deque = []
        self.size = 0
    def push(self, value):
        self.deque.append(value)
        self.size += 1
    def push_front(self, value):
        self.deque.insert(0, value)
        self.size += 1
    def pop(self):
        if self.is_empty(): 
            return None
        self.size -= 1
        return self.deque.pop()
    def pop_front(self):
        if self.is_empty(): 
            return None
        self.size -= 1
        return self.deque.pop(0)
    def peek(self):
        return self.deque[self.size - 1]
    def peek_front(self):
        return self.deque[0]
    def is_empty(self):
        return self.size == 0
```
